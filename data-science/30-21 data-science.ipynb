## pandas df.filter()

```python
import pandas as pd
import numpy as np
```

```python
data=pd.DataFrame({'one':np.arange(10),'two':np.arange(10,20),'three':np.arange(20,30)})
```

```python
data
```

```python
data1=data.filter(regex='e$',axis=1) # 选出以e结尾为列名的列
```

```python
data1
```

```python
data2=data.filter(regex='^t',axis=1)
```

```python
data2
```

```python
data3=data.filter(regex='.*h.*',axis=1)
```

```python
data3
```

其中的regex表示的是使用正则表达式的形式来进行条件选择需要的列或者行，如果是行的话，那么axis=0

```python
data4=data.filter(items=['one'])
```

```python
data4
```

filter的参数还可以是列名，这样其实跟直接使用括号一样，如果使用这种方式的话，还不如用括号的方式呢
